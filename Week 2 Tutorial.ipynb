```python
import numpy as np
```

```python
my_arr = np.arange(10)
print("my_arr: ", my_arr)
my_arr[4:7] = 25
print("my_arr: ", my_arr)
arr_slice = my_arr[4:7]

# Change the first element of arr_slice to -1
arr_slice[0]= -1

print("sliced arr: ", arr_slice)
print("my_arr: ", my_arr)
```

# What do you notice? Propose a way to reassign arr_slice with new value without modifying my_arr

changing the arr_slice elements will change my_arr elements as well.
instead of assigning my_arr to arr_slice, we can use np.copy()

```python
my_arr = np.arange(10)
print("my_arr: ", my_arr)
my_arr[4:7] = 25
print("my_arr: ", my_arr)
arr_slice = np.copy(my_arr[4:7])

# Change the first element of arr_slice to -1
arr_slice[0]= -1

print("sliced arr: ", arr_slice)
print("my_arr: ", my_arr)
```

```python
import matplotlib.pyplot as plt
```

```python
gray = np.zeros((200, 200), dtype="uint8")

plt.imshow(gray, cmap = plt.cm.gray)
plt.title("grayscale black image")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
white = 255 * np.ones((200, 200))
white = np.uint8(white)

plt.imshow(white, cmap = plt.cm.gray, vmin=0, vmax=255)
plt.title("grayscale white image")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Allow user to set the dimension of the image and the circle
Xdimension = int(input("Width of the image: "))
Ydimension = int(input("Height of the image: "))
radius = int(input("Radius of the circle: "))

gray = np.zeros((Ydimension, Xdimension), dtype="uint8")
#Create circle
circle = plt.Circle((Xdimension/2, Ydimension/2),radius, color='white')

fig, ax = plt.subplots() # note we must use plt.subplots, not plt.subplot
ax.add_patch(circle)

plt.imshow(gray, cmap = plt.cm.gray)
plt.title("Image with circle")
plt.xticks([]), plt.yticks([])
plt.show()
```
